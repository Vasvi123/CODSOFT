```python
import csv
def CreateItem():
   with open("Contact.csv",'w',newline="") as f:
        cv = csv.writer(f,delimiter =',')
        cv.writerow(['Name','Phone Number','EmailId','Address'])
        while True:
           Name = input("Enter your name :")
           Phone = input("Enter your phone number : ")
           email = input("Enter your email id : ")
           address = input("Enter your address : ")
           cv.writerow([Name,Phone,email,address])
           choice = input("Enter more? (y/Y/n/N) : ")
           if choice in "nN":
                break
def DisplayItem():
   try:
       with open("Contact.csv") as f:
           cv = csv.reader(f,delimiter = ',')
           for c in cv:
              print(c)
   except:
       print("File not found")
def AddItem():
   with open("Contact.csv",'a',newline="") as f:
          cv = csv.writer(f,delimiter =',')
          while True:
              Name = input("Enter your name :")
              Phone = input("Enter your phone number : ")
              email = input("Enter your email id : ")
              address = input("Enter your address : ")
              cv.writerow([Name,Phone,email,address])
              choice = input("Enter more? (y/Y/n/N) : ")
              if choice in "nN":
                  break
def SearchItem():
   Name1 = input("Enter the Name to be searched :")
   try:
       with open("Contact.csv",newline="") as f:
             cv = csv.reader(f,delimiter = ',')
             for c in cv:
                if c[0] == Name1:
                      print(c)
                      break
             else:
                      print(Name1, " not found in the file")
   except:
        print("File not found")
def UpdateItem():
   try:
      with open("Contact.csv",'r+') as f:
           Name2 = input("Enter the Name to be updtaed: ")
           cv = csv.reader(f,delimiter = ',')
           L=[]
           changed = False
           for c in cv:
               if c[0]==Name2:
                      print("Name:",c[0],"Phone number:",c[1])
                      c[1]= int(input("Enter new phone number:"))
                      changed = True
               L.append(c)
           if changed == False:
               print(Name2, " not present in the file")
           else:
               f.truncate(0)
               f.seek(0)
               cw = csv.writer(f)
               cw.writerows(L)
   except :
          print("File not found ")
def DeleteItem():
       i=int(input("Enter the index of the row to be deleted:"))
       with open("Contact.csv", 'r') as file:
          rows= list(csv.reader(file))
          if  i< 0 or i>= len(rows):
               print("Invalid row index.")
               return
          del rows[i]
       with open("Contact.csv", 'w', newline='') as file:
          writer = csv.writer(file)
          writer.writerows(rows)

while True:
      print("1. Create a file  ")
      print("2. Display the file ")
      print("3. Add contact details into the file ")
      print("4. Search the file  ")
      print("5. Update contact details in the file  ")
      print("6. Delete contact details  ")
      print("7. Quit the program (Q/q) ")
      choice = int(input("Enter your choice(1/2/3/4/5/6) : "))
      if choice==1:
          CreateItem()
      elif choice==2:
          DisplayItem()
      elif choice==3:
          AddItem()
      elif choice==4:
          SearchItem()
      elif choice==5:
          UpdateItem()
      elif choice==6:
          DeleteItem()
      elif choice==7:
           break
      else:
           break
```
